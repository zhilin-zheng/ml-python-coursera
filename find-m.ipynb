```python
import numpy as np

x = np.random.normal(size=(2, 1000))
y = np.random.normal(size=(1, 1000))
print(type(x))
print(x.shape)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = x.T
print(x)
y = y.T
print(y)
regr.fit(x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```
